```python
# Connection to Postgres Yelp Database - Table: restaurant_reviews_final
conn = psycopg2.connect(database="yelp", user="postgres",
                        password='password', host="localhost", port="5432")
query = '''
        SELECT *
        FROM restaurant_reviews_final
        LIMIT 10000
        ;
        '''
df = pd.read_sql(query, conn)
# print(df.info())
```

### Combine Columns Where Individual Columns are not Useful

```python
# Combining sub-targets
# Consider weighting them separately in the future.
test['review_total_positives'] = test[['review_useful', 'review_funny', 'review_cool']].sum(axis=1)
test.head(5)
```

```python
test.drop(labels=['review_useful', 'review_funny', 'review_cool'], axis=1, inplace=True)
```

```python
test['user_total_positives'] = test[['user_useful', 'user_funny', 'user_cool']].sum(axis=1)
test.drop(labels=['user_useful', 'user_funny', 'user_cool'], axis=1, inplace=True)
test.head()
```

```python
compliment_columns = ['user_compliment_hot', 'user_compliment_more', 
                      'user_compliment_profile', 'user_compliment_cute',
                      'user_compliment_list', 'user_compliment_note',
                      'user_compliment_plain', 'user_compliment_cool',
                      'user_compliment_funny', 'user_compliment_writer',
                      'user_compliment_photos']
test['user_total_compliments'] = test[compliment_columns].sum(axis=1)
test.drop(labels=compliment_columns, axis=1, inplace=True)
test.head()
```

### Split List Columns Into Useful Columns

```python
def count_friends(x):
    friend_count = 0 
    if x in ['None', None, '']:
        friend_count = 0
    else:
        y = x.split(',')
        friend_count = len(y)
    return friend_count


test['user_friends_count'] = test.user_friends.apply(count_friends)
test.head(10)
```

```python
def count_elite(x):
    elite_count = 0 
    if x in ['None', None, '']:
        elite_count = 0
    else:
        y = x.split(',')
        elite_count = len(y)
    return elite_count

test['user_elite_count'] = test.user_elite.apply(count_elite)
test.head(10)
```

```python
def most_recent_elite(x):
    most_recent = 0 
    if x in ['None', None, '']:
        most_recent = 0
    else:    
        y = x.split(',')
        most_recent = max(y)
    return most_recent

test['user_elite_most_recent'] = test.user_elite.apply(most_recent_elite)
test.head(10)
```

```python
test.head(5)
```

```python
def count_checkins(x):
    checkin_count = 0 
    if x in ['None', None, '']:
        checkin_count = 0
    else:
        y = x.split(',')
        checkin_count = len(y)
    return checkin_count

test['restaurant_checkin_count'] = test.restaurant_checkins.apply(count_checkins)
test.head(10)
```

```python
test.drop(labels=['restaurant_categories', 'user_friends', 'user_elite', 'restaurant_checkins'], axis=1, inplace=True)
test.info()
```
